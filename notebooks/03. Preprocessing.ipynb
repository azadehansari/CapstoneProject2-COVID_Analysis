```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv('../data/features_target_unscaled.csv')

df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head()
```

```python
X = df.drop(columns = ['DIED'])
Y = df['DIED']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state= 42)
```

```python
df.columns
```

```python
y_train =y_train.map({1 : 0, 2: 1})
y_test =y_test.map({1 : 0, 2: 1})
```

```python
encode_cols = ['USMER', 'MEDICAL_UNIT', 'SEX', 'PATIENT_TYPE', 'INTUBED', 'PNEUMONIA',
       'PREGNANT', 'DIABETES', 'COPD', 'ASTHMA', 'INMSUPR',
       'HIPERTENSION', 'CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO',
       'ICU', 'COVID']

encode_data = x_train[encode_cols]

encoder = OneHotEncoder(sparse=False, drop='first')

tr_encode = encoder.fit_transform(encode_data)
```

```python
df_encode_tr = pd.DataFrame(tr_encode, columns= encoder.get_feature_names_out(encode_cols))

df_encode_tr.head()
```

```python
x_train = x_train.reset_index()
df_encode_tr = df_encode_tr.reset_index()

tr_df = pd.concat((x_train['AGE'], df_encode_tr), axis=1) 
tr_df.head()
```

```python
tr_df.drop(columns = ['index'], inplace=True)
tr_df.head()
```

# Encode Test set

```python
encode_data_ts = x_test[encode_cols]

ts_encode = encoder.transform(encode_data_ts)

df_encode_ts = pd.DataFrame(ts_encode, columns= encoder.get_feature_names_out(encode_cols))

df_encode_ts.head()
```

```python
x_test = x_test.reset_index()
df_encode_ts = df_encode_ts.reset_index()

ts_df = pd.concat((x_test['AGE'], df_encode_ts), axis=1) 
ts_df.head()
```

```python
ts_df.drop(columns = ['index'], inplace=True)
```

```python
ts_df.head()
```

# Scale Age 

```python
scaler = MinMaxScaler()

tr_df['AGE'] = scaler.fit_transform(tr_df[['AGE']])

tr_df.head()
```

```python
ts_df['AGE'] = scaler.transform(ts_df[['AGE']])
ts_df.head()
```

```python
tr_df = tr_df.reset_index()
y_train = y_train.reset_index()

train = pd.concat([tr_df, y_train], axis = 1)
train.drop(columns= ['index'], inplace=True)
train.head()
```

```python
ts_df = ts_df.reset_index()
y_test = y_test.reset_index()

test = pd.concat([ts_df, y_test], axis = 1)
test.drop(columns= ['index'], inplace=True)
test.head()
```

```python
train.to_csv('../data/train.csv')
test.to_csv('../data/test.csv')
```
